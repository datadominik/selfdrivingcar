## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

---


## Camera calibration
Calibration code used as provided in the example notebook.

```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
%matplotlib inline

# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('camera_cal/calibration*.jpg')

# Step through the list and search for chessboard corners
for fname in images:
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)

    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)

ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)

def undistort(img, mtx, dist):
    return cv2.undistort(img, mtx, dist, None, mtx)

undist = undistort(img, mtx, dist)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.axis('off')
ax2.axis('off')

ax1.imshow(img)
ax2.imshow(undist)
```

## Binary thresholding

### Sobel

```python
test_img = 'test_images/straight_lines1.jpg'

def sobel_filter(img, min_thresh=50, max_thresh=255):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    sobel = np.uint8(sobel*255./np.max(sobel))
    
    binary = np.zeros_like(sobel)
    binary[(sobel > min_thresh) & (sobel < max_thresh)] = 1
    
    return binary

img = cv2.imread(test_img)
img_sob = sobel_filter(img)


fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.axis('off')
ax2.axis('off')

ax1.imshow(img)
ax2.imshow(img_sob)
```

### HLS filtering

```python
def hls_filter(img, min_thresh=150, max_thresh=255):
    hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    s_channel = hls[:,:,2]
    
    binary = np.zeros_like(s_channel)
    binary[(s_channel > min_thresh) & (s_channel < max_thresh)] = 1
    
    return binary

img = cv2.imread(test_img)
img_hls = hls_filter(img)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.axis('off')
ax2.axis('off')

ax1.imshow(img)
ax2.imshow(img_hls)
```

### Combined filtering

```python
img_comb = np.asarray([img_hls,img_sob]).mean(axis=0)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.axis('off')
ax2.axis('off')

ax1.imshow(img)
ax2.imshow(img_comb)
```

## Perspective transform

```python
x_offset = 500
y_offset = 300
width = img.shape[1]
height = img.shape[0]

src = np.float32([[width-x_offset,height-y_offset],[width,height],[0,height],[0+x_offset,height-y_offset]])
dst = np.float32([[width,height-y_offset],[width,height],[0,height],[0,height-y_offset]])

w,h = img.shape[1],img.shape[0]
x,y = 0.5*w, 0.8*h

src = np.float32([[200.,h],
              [450.,550.],
              [840.,550.],
              [1100.,h]])

dst = np.float32([[(w-x)/2.,h],
              [(w-x)/2.,0.8*h],
              [(w+x)/2.,0.8*h],
              [(w+x)/2.,h]])


plt.imshow(img)
plt.axis('off')

for corner in src:
    plt.plot(corner[0],corner[1],".",color="red")
    
```

```python
M = cv2.getPerspectiveTransform(src, dst)
Minv = cv2.getPerspectiveTransform(dst, src)

img_size = (img.shape[1],img.shape[0])

def warp_image(img):
    return cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)

img_warp = warp_image(img)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.axis('off')
ax2.axis('off')

ax1.imshow(img)
ax2.imshow(img_warp)
```

## Video Processing

```python
class Line():
    def __init__(self):
        self.recent = collections.deque(maxlen=20)
        self.best_x = None     
        self.best_fit = None  
        self.radius_of_curvature = None 
```

```python
import collections

left_line = Line()
right_line = Line()


def process_frame(img):
    img_tmp = np.copy(img)
    
    ### Constants
    n_windows = 9
    margin = 100
    min_pix = 50
    
    ### Undistort, binarize and warp image
    img_undist = undistort(img_tmp, mtx, dist)   
    img_hls = hls_filter(img_undist)
    img_sob = sobel_filter(img_undist)

    binary = np.asarray([img_hls,img_sob]).mean(axis=0)
    binary_warped = warp_image(binary)
    
    ### Lane detection
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    
    window_height = np.int(binary_warped.shape[0]/n_windows)
    
    # find two peaks in histogram, one per half
    max_left = np.argmax(histogram[:img.shape[1]//2])
    max_right = img.shape[1]//2+np.argmax(histogram[img.shape[1]//2:img.shape[1]])

    non_zero = binary_warped.nonzero()
    non_zero_y = np.array(non_zero[0])
    non_zero_x = np.array(non_zero[1])

    left_x = max_left
    right_x = max_right

    left_lane_idx = []
    right_lane_idx = []

    for window in range(n_windows):
        win_y_top = binary_warped.shape[0]-window*window_height
        win_y_bottom = binary_warped.shape[0]-(window+1)*window_height

        win_left_left = left_x - margin
        win_left_right = left_x + margin

        win_right_left = right_x - margin
        win_right_right = right_x + margin

        left_idx = ((non_zero_y >= win_y_bottom) & (non_zero_y < win_y_top) & (non_zero_x >= win_left_left) & (non_zero_x < win_left_right)).nonzero()[0]
        right_idx = ((non_zero_y >= win_y_bottom) & (non_zero_y < win_y_top) & (non_zero_x >= win_right_left) & (non_zero_x < win_right_right)).nonzero()[0]

        left_lane_idx.append(left_idx)
        right_lane_idx.append(right_idx)

        if len(left_idx) > min_pix:
            left_x = int(np.mean(non_zero_x[left_idx]))
        if len(right_idx) > min_pix:
            right_x = int(np.mean(non_zero_x[right_idx]))

    left_lane_idx = np.concatenate(left_lane_idx)
    right_lane_idx = np.concatenate(right_lane_idx)

    # Extract left and right line pixel positions
    left_x = non_zero_x[left_lane_idx]
    left_y = non_zero_y[left_lane_idx] 
    right_x = non_zero_x[right_lane_idx]
    right_y = non_zero_y[right_lane_idx] 

    # Fit a second order polynomial to ea
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    
    if((len(left_x)> 0) & (len(right_x)> 0)):
        left_fit = np.polyfit(left_y, left_x, 2)
        right_fit = np.polyfit(right_y, right_x, 2)
    
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
        
        # Append to FiFo 
        left_line.recent.append(left_fitx)
        right_line.recent.append(right_fitx)
        
    left_fitx = np.asarray(left_line.recent).mean(axis=0)
    right_fitx = np.asarray(right_line.recent).mean(axis=0)
    
    # Values and parts of code from the lecture
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    y_max = np.max(ploty)
    
    left_fit_cr = np.polyfit(ploty*ym_per_pix, left_fitx*xm_per_pix, 2)
    right_fit_cr = np.polyfit(ploty*ym_per_pix, right_fitx*xm_per_pix, 2)
    
    left_curve_rad = ((1 + (2*left_fit_cr[0]*y_max*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curve_rad = ((1 + (2*right_fit_cr[0]*y_max*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    
    curve_rad = int((left_curve_rad + right_curve_rad)/2)
    if(left_line.radius_of_curvature != None):
        if(curve_rad > left_line.radius_of_curvature):
            tmp_1 = curverad
            tmp_2 = left_line.radius_of_curvature
        else:
            tmp_2 = curve_rad
            tmp_1 = left_line.radius_of_curvature            
        if(tmp_1/tmp_2 > 5):
            left_fitx = left_line.best_x
            right_fitx = right_line.best_x
            curve_rad = left_line.radius_of_curvature
        
    # Create an image to draw the lines on
    warp_zero = np.zeros_like(binary_warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    
    # projection is used to show the warped image as well
    projection =  cv2.addWeighted(warp_image(img), 1, color_warp, 0.3, 0)
    projection = cv2.resize(projection, (projection.shape[1]//5,projection.shape[0]//5))
    
    img_sob = cv2.resize(img_sob, (projection.shape[1],projection.shape[0]))
    img_hls = cv2.resize(img_hls, (projection.shape[1],projection.shape[0]))
    
    new_warp = cv2.warpPerspective(color_warp, Minv, (img.shape[1], img.shape[0])) 
    
    result = cv2.addWeighted(img, 1, new_warp, 0.3, 0)
    result[50:50+projection.shape[0],50:50+projection.shape[1]] = projection

    # Calculate position with respect to center
    camera_pos = img.shape[1]/2

    lane_pos = (right_fitx[719] + left_fitx[719])/2

    offset = abs(camera_pos - lane_pos) * 3.7/700
    
    cv2.putText(result,'curvature: {}m'.format(curve_rad),(350,75),cv2.FONT_HERSHEY_SIMPLEX, 1,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(result,'offset: {:.2f}m'.format(offset),(350,125),cv2.FONT_HERSHEY_SIMPLEX, 1,(255,255,255),2,cv2.LINE_AA)
    
    left_line.best_x = left_fitx
    right_line.best_x = right_fitx
    
    left_line.radius_of_curvature = curve_rad
    right_line.radius_of_curvature = curve_rad
    
    return result


img = cv2.imread('test_images/test1.jpg')
result = process_frame(img)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.axis('off')
ax2.axis('off')

ax1.imshow(img)
ax2.imshow(result)
```
