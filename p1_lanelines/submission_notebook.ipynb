## Helper Functions (I didn't change)

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, alpha=0.8, beta=1., gamma=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, alpha, img, beta, gamma)
````

## Changed helper functions & new functions
Here are the functions I changed or added to achieve the video results.

````python
def split_lines(lines):
    """
    NOTE: this function is used in order to differentiate between left and right lanes. In order
    to do so I calculate the slope with given points and set a hard slope threshold (-0.25,0.25)
    that worked good to deal with outliers.
    """
    left_lines = []
    right_lines = []
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            #calculate slope m of given points
            m = (y2-y1)/(x2-x1)
            if (m >= 0.25):
                right_lines.append(line)
            elif (m <= -0.25):
                left_lines.append(line)
        
    return left_lines, right_lines

def extrapolate_lines(img, lines):
    """
    NOTE: this function is used to smoothen/ extrapolate the lines we receive with canny + hough transform.
    
    """
    X_train = np.array([np.vstack(lines)[:,0],np.vstack(lines)[:,2]]).flatten()
    y_train = np.array([np.vstack(lines)[:,1],np.vstack(lines)[:,3]]).flatten()
    
    m, intercept, _, _, _ = stats.linregress(X_train,y_train)
    
    if(m<=-0.25):
        min_x = int((img.shape[0]-intercept)/m)
        max_x = int((img.shape[0]/1.65-intercept)/m)
    elif(m>=0.25):
        min_x = int((img.shape[0]/1.65-intercept)/m)
        max_x = int((img.shape[0]-intercept)/m)
        
    line = []
    line.append([np.array([min_x,int(m*min_x+intercept),max_x,int(m*max_x+intercept)])])
    return line

def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    # the only thing I changed here is to directly include my extrapolate lines function
    # and changing the thickness of the line. Like this we get both a single line for left and one for right.
    
    lines = extrapolate_lines(img,lines)       
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    
    # split hough lines into left and right 
    left_lines, right_lines = split_lines(lines)
    
    # draw left and right lane in different color for visualization purposes
    draw_lines(line_img, left_lines, color=[255, 0, 0])
    draw_lines(line_img, right_lines, color=[0, 0, 255])
    
    return line_img, lines
````

### pre-work for building the pipeline
in the next cell I basically built the foundation for the pipeline later. I called the functions described above in the following order: 
- load image and make a copy
- convert to grayscale
- blur the image with kernel=7 to remove unnecessary details/ noise
- apply canny filter
- define vertices to define my region of interest
- apply hough transform (together with line seperation and extrapolation)
- combined lines with original image

All parameters in canny, gaussian_blur, region_of_interest and hough_transform are currently hand tuned. This will be a shortcoming in real life situations since we will be heavily overfitted to our testset here. Nevertheless for the images and videos later it works quite well.

````python
fig, axs = plt.subplots(2,3, figsize=(20, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .3, wspace=.05)

axs = axs.ravel()

for i in range(6):
    img_original = plt.imread(test_files[i])
    
    img = np.copy(img_original)
    img = grayscale(img)
    img = gaussian_blur(img,7)

    img = canny(img,190,200)

    vertices = np.array([[(0,img.shape[0]), (img.shape[1]/2,int(img.shape[0]/1.7)),(img.shape[1]/2,int(img.shape[0]/1.7)),(img.shape[1],img.shape[0])]],dtype="int32")
    img = region_of_interest(img,vertices)

    img, lines = hough_lines(img, 1, np.pi/180, 20, 80, 200)

    result = weighted_img(img,img_original)
    
    axs[i].imshow(result)
    axs[i].set_title('image {}'.format(i))

fig.savefig("examples/transformed.jpg")
````

````python
def process_image(img_original):
    """
    NOTE: this processing pipeline is exactly the same as described above. All function parameters are hand tuned and most
    likely overfitted to this training set.
    """
    img = np.copy(img_original)
    img = grayscale(img)
    img = gaussian_blur(img,7)
    
    img = canny(img,190,200)
    
    vertices = np.array([[(0,img.shape[0]), (img.shape[1]/2,int(img.shape[0]/1.7)),(img.shape[1]/2,int(img.shape[0]/1.7)),(img.shape[1],img.shape[0])]],dtype="int32")
    img = region_of_interest(img,vertices)

    img, lines = hough_lines(img, 1, np.pi/180, 20, 80, 200)

    result = weighted_img(img,img_original)
    return result
````
